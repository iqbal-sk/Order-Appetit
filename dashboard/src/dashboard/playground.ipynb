```python
def create_analysis_tasks(user_query):
    schema_analysis_task = Task(
        description=f"Analyze MongoDB collections and identify relevant document patterns for: {user_query}",
        expected_output="Dictionary mapping query components to collection structures",
        agent=schema_analyzer
    )
    
    query_building_task = Task(
        description="Generate optimized MongoDB aggregation pipeline or query based on schema analysis",
        expected_output="Executable MongoDB query or aggregation pipeline",
        agent=query_builder,
        context=[schema_analysis_task]
    )
    
    data_analysis_task = Task(
        description="Analyze MongoDB query results and generate insights",
        expected_output="Dictionary containing analysis results and key findings",
        agent=data_analyst,
        context=[query_building_task, schema_analyzer]
    )
    
    formatting_task = Task(
        description="Format MongoDB analysis results for Chainlit display",
        expected_output="Formatted response with visualizations if applicable",
        agent=response_formatter,
        context=[data_analysis_task]
    )
    
    return [schema_analysis_task, query_building_task, data_analysis_task, formatting_task]
```
